```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n_eq = 22

# eq = np.zeros(22, dtype = np.ndarray)

# for i in range(0,n_eq):
#     eq[i] = 
    
eq_list = [np.squeeze(pd.read_csv(f"GMs\EQ{i+1}.txt").values) for i in range(n_eq)]
```

```python
shapes = [eq_list[i].shape for i in range(n_eq)]
```

```python
ndata = max(shapes)[0]

eq = np.zeros((n_eq, ndata))
for i in range(n_eq):
    eq[i] = np.concatenate((eq_list[i], np.zeros(ndata - len(eq_list[i]))), axis = 0)
```

```python
print(shapes)
print(np.max(shapes))
print(eq.shape)
# print(n_data)
```

```python
plt.plot(eq[4]);
plt.xlabel('time');
plt.ylabel('g');
```

```python
# ndata = eq.shape[0]
```

```python
# implicit
gama = 0.5
beta = 0.25

# explicit
# gama = 0.5
# beta = 1/6
```

```python
n_period = 100
ksi = 0.05
m = 1.
delta_t = 0.01

period = np.linspace(0.0001,5,n_period)

u = np.zeros((n_period,ndata,3,n_eq), dtype=float)

k = 4 * np.pi * m / np.square(period)
k = k[:,np.newaxis]
c = 2 * m * ksi * np.sqrt(k/m)


p = eq * m * 9.81

u[:,0,2,:] = (p[:,0] - c * u[:,0,1,:] - k * u[:,0,0,:]) / m
a1 = (m / (beta * (delta_t ** 2))) + (gama * c) / (beta * delta_t)
a2 = (m / (beta * delta_t)) + ((gama/beta) - 1) * c
a3 = ((1/(2 * beta)) - 1) * m + delta_t * ((gama/(2 * beta)) - 1) * c
k_star = k + a1
```

```python
for i in range(0,ndata-1):
    p_star = p[:,i+1] + a1 * u[:,i,0,:] + a2 * u[:,i,1,:] + a3 * u[:,i,2,:]
    u[:,i+1,0,:] = p_star / k_star
    u[:,i+1,1,:] = gama * (u[:,i+1,0,:]-u[:,i,0,:]) / (beta * delta_t) + u[:,i,1,:] * (1 - gama/beta) + delta_t * u[:,i,2,:] * (1 - gama/(2*beta))
    u[:,i+1,2,:] = (u[:,i+1,0,:]-u[:,i,0,:]) / (beta * (delta_t ** 2)) - u[:,i,1,:] / (beta * delta_t) - u[:,i,2,:] * ((1 / (2*beta)) - 1)
```

```python
plt.plot(u[1,:,1,2]);
plt.xlabel('time');
plt.ylabel('g');
```

```python
plt.plot(np.max(np.abs(u[:,:,0,2]),axis=1));
plt.xlabel('time');
plt.ylabel('sd');
```
